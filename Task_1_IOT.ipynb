## > Importing required libraries

```python
import keras_ocr as kr
import matplotlib.pyplot as plt
```

```python
# Automatically downloads pretrained weights for the detector and recognizer
pipeline=kr.pipeline.Pipeline()
```

## > Importing Images

```python
image_ocr=[
           kr.tools.read(url) for url in['https://images.unsplash.com/photo-1568849676085-51415703900f?ixid=MnwxMjA3fDB8MHxwaG90by1wYWdlfHx8fGVufDB8fHx8&ixlib=rb-1.2.1&auto=format&fit=crop&w=334&q=80',
                                         'https://images.unsplash.com/photo-1495001258031-d1b407bc1776?ixid=MnwxMjA3fDB8MHxzZWFyY2h8NTB8fHF1b3RlfGVufDB8fDB8fA%3D%3D&ixlib=rb-1.2.1&auto=format&fit=crop&w=500&q=60',
                                         'https://static.wikia.nocookie.net/listofdeaths/images/0/07/The_Dark_Knight_poster.jpg/revision/latest/scale-to-width-down/250?cb=20201107115458'
                                         ]
]
```

```python
len(image_ocr)
```

## > Displaying the Image

```python
plt.figure(figsize=(10,10))
plt.imshow(image_ocr[0])
```

```python
plt.figure(figsize=(10,10))
plt.imshow(image_ocr[1])
```

```python
plt.figure(figsize=(10,10))
plt.imshow(image_ocr[2])
```

```python
prediction_groups=pipeline.recognize(image_ocr)
```

## > Result
### Get predictions

```python
fig, axs = plt.subplots(nrows=len(image_ocr), figsize=(15, 15))
for ax, image, predictions in zip(axs, image_ocr, prediction_groups):
    kr.tools.drawAnnotations(image=image, predictions=predictions, ax=ax)
```
